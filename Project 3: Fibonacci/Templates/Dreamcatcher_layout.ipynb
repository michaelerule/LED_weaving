```python
N = 13
K = N*2
h = 2*pi/K
c = exp(1j*linspace(pi,-pi,K+1))
#c2 = exp(1j*linspace(0,2*pi,1000))

NBASIS = N
colors = array([VIOLET,MAUVE,CRIMSON,RUST,OCHRE,CHARTREUSE,MOSS,TURQUOISE,AZURE,VIOLET])
ncolor = len(colors)
riley  = matplotlib.colors.LinearSegmentedColormap.from_list('riley',colors)
colors = riley(linspace(0,(ncolor-1)/ncolor,NBASIS))[:,:3]

figure(figsize=(14,14))
force_aspect()
noxyaxes()

d  = 1

for i,t in enumerate(c[:-1]):
    z = t*(c2+1.01)
    r = abs(z)
    no = c2.real<-0.05
    #v = exp(r*2) + 1/r
    #v = 60/(1+exp(-r*4)) + exp(3*r)
    v = 1
    #r = r*v+d
    r = r**0.9
    r -= np.min(r[~no])
    r += np.max(r)*0.25
    z = z/abs(z)*r
    z[no]=NaN
    plot(*c2p(z),color=colors[i%N],lw=3)
    
    s = '%d'%(i+1) if i<N else "%d'"%(i-N+1)
    text(*c2p(t)*max(r*1.1),s,fontsize=30,ha='center',va='center')
    
savefigure('Dreamcatcher_%d'%N)
```
