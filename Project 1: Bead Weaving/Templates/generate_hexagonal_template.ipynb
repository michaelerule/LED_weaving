# Generate hexagonal template

```python
A4sizecm  = array([21,29.7])
margincm  = 2
cmperinch = 2.54
figsizecm = A4sizecm#-2*margincm
beadlenmm = 8.5
padlenmm  = 0.5
overlapmm = 3.0
gapmm     = 0.5
netlenmm  = beadlenmm+padlenmm*2+overlapmm+gapmm
leadlenmm = netlenmm+overlapmm-gapmm
triangleh = sqrt(3)/2
linesepmm = netlenmm*triangleh
print('Net  length is',netlenmm)
print('Lead length is',leadlenmm)
figure(figsize=figsizecm/cmperinch,facecolor=(1,1,1,0),frameon=False)
subplot(111)
noxyaxes()
for u in exp(1j*linspace(0,2*pi,4)[:-1]):
    
    shift = 1j*pi/3
    
    spacing = arange(-40,40)*0.1*linesepmm*1j
    grid = array([spacing-150,spacing+150]).T
    for u in exp(shift+1j*linspace(0,2*pi,4)[:-1]):
        for p1,p2 in grid*u:
            plot([p1.real,p2.real],[p1.imag,p2.imag],color=(0.5,0.5,0.5),lw=0.6)
            
    factor = netlenmm
    spacing = arange(-40,40)*0.1*factor*1j
    grid = array([spacing-150,spacing+150]).T
    for u in exp(shift+1j*pi/6+1j*linspace(0,2*pi,4)[:-1]):
        for p1,p2 in grid*u:
            plot([p1.real,p2.real],[p1.imag,p2.imag],color=(0.5,0.5,0.5),lw=0.4)
            
    factor = netlenmm*.5
    spacing = arange(-40,40)*0.1*factor*1j
    grid = array([spacing-150,spacing+150]).T
    for u in exp(shift+1j*pi/6+1j*linspace(0,2*pi,4)[:-1]):
        for p1,p2 in grid*u:
            plot([p1.real,p2.real],[p1.imag,p2.imag],color=(0.5,0.5,0.5),lw=0.2)

mx,my = margincm/A4sizecm[0],margincm/A4sizecm[1]
subplots_adjust(mx,my,1-mx,1-my)
xlim(0,figsizecm[0]-2*margincm)
ylim(0,figsizecm[1]-2*margincm)

savefig('hexgrid.ps',facecolor=(1,1,1,0),edgecolor=(1,1,1,0),
        orientation='portrait', papertype='a4', format='ps',
        transparent=True, pad_inches=0,
        frameon=False)
```
