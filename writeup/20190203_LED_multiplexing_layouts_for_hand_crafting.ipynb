# LED multiplexing for hand-crafting

Applications: low-budget manual fabrication of LED matrix displays and marquees, as well as e-textiles: sewn and woven LED displays as well as bead-weaving LED displays. 

## LED multiplexing

```python
N = 7
size = 200
lw = 1.5
figsize=(2.5,2.5)

figure(figsize=figsize)
gridxy = meshgrid(range(N),range(N))
scatter(*gridxy,s=size,facecolor=(1,1,1,0),edgecolor=BLACK,clip_on=False)
noxyaxes()

for i in range(N):
    text(-1,i,'%d'%(N-i),fontsize=12,va='center')
    text(i,N+0.1,'%d'%(i+1+N),fontsize=10,ha='center',va='top')

for i in range(N):
    text(-1.75,i,'$\ominus$' if i==N-2 else '$\circ$',fontsize=14,va='center',ha='center')

on = (0,1,3,6,7,8)
for x in range(N):
    text(x,N+0.75,'$\oplus$' if x in on else '$\circ$',fontsize=14,va='center',ha='center')
    if x in on:
        scatter([x],N-2,s=size,facecolor=MAUVE,edgecolor=BLACK,clip_on=False)
        plot([x,x],[-.5,N-0.5],color=OCHRE,lw=lw)
    
y1 = N-1.5
y2 = N-2.5
px = array([0,N,N,0,0])-0.5
py = [y1,y1,y2,y2,y1]
plot([-0.5,N-0.5],[N-2,N-2],lw=lw,color=TURQUOISE)

tight_layout()
title('Multiplexing',pad=30)
ylabel('Cathodes',labelpad=22)
xlabel('Anodes',labelpad=0)
saveit()
```
